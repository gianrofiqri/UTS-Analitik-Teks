Fungsi Pra-Pemrosesan

```python
def preprocess_text(text):
    if not isinstance(text, str):
        return ""

    # 1. Lowercasing
    text = text.lower()

    # 2. Hapus Tanda Baca, Karakter Khusus, Angka, URL, dan Mention
    # Hapus URL
    text = re.sub(r'&[a-z]+;', ' ', text)
    # Hapus mention
    text = re.sub(r'http\S+|www\S+|https\S+', ' ', text, flags=re.MULTILINE)
    text = re.sub(r'\@\w+', ' ', text)
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)

    # 3. Tokenisasi
    tokens = word_tokenize(text)

    # 4. Penghapusan Stopwords dan 5. Stemming (dilakukan bersamaan)
    cleaned_tokens = []
    for token in tokens:

        if token not in stop_words and len(token) > 1:
            cleaned_tokens.append(token)

    # Gabungkan kembali token menjadi satu string bersih
    return " ".join(cleaned_tokens)
```

# 1. TF-IDF (Term Frequency-Inverse Document Frequency)
Penjelasan Konsep:


TF-IDF adalah sebuah metode statistik untuk memberi bobot (skor) pada sebuah kata. Tujuannya adalah untuk mengukur seberapa penting sebuah kata dalam sebuah dokumen relatif terhadap seluruh kumpulan dokumen (korpus).


*   Term Frequency (TF): Seberapa sering sebuah kata muncul dalam satu dokumen.Contoh: Jika kata "ganoderma" muncul 3 kali di Tweet A, TF-nya tinggi untuk Tweet A.
*   Inverse Document Frequency (IDF): Seberapa langka sebuah kata di seluruh korpus.Contoh: Kata "mari" (jika tidak dihapus) akan muncul di hampir semua dokumen, sehingga nilai IDF-nya sangat rendah (tidak penting).Contoh: Kata "ganoderma" mungkin hanya muncul di 5% dokumen. Ini membuatnya langka dan spesifik, sehingga nilai IDF-nya sangat tinggi (penting).



Skor Akhir: $TF-IDF = TF \times IDF$Sebuah kata akan mendapat skor tinggi jika ia sering muncul di dokumen itu (TF tinggi) TAPI jarang muncul di dokumen lain (IDF tinggi). Ini adalah cara yang bagus untuk menemukan kata kunci dari sebuah dokumen.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

if 'df' not in locals() or 'text_cleaned' not in df.columns:
    print("Error: DataFrame 'df' atau kolom 'text_cleaned' tidak ditemukan.")
else:
    # Ambil kumpulan teks dari DataFrame
    # .fillna('') untuk mengatasi jika ada baris kosong setelah pra-pemrosesan
    corpus = df['text_cleaned'].fillna('')

    # 1. Inisialisasi TfidfVectorizer
    # max_features=5000 berarti hanya ambil 5000 kata dengan skor TF-IDF tertinggi
    # Ini membantu mengurangi noise dan ukuran matriks
    tfidf_vectorizer = TfidfVectorizer(max_features=5000)

    # 2. Latih (fit) dan Transformasi (transform)
    # .fit_transform() melakukan dua hal:
    # 1. fit: Mempelajari vocabulary (semua kata unik) dan menghitung skor IDF.
    # 2. transform: Mengubah setiap teks menjadi vektor TF-IDF.
    tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)

    # 3. Tampilkan hasilnya
    print("## Hasil Representasi TF-IDF ##")
    print(f"Bentuk (shape) dari matriks TF-IDF: {tfidf_matrix.shape}")
    print(f"Artinya: {tfidf_matrix.shape[0]} dokumen/tweet")
    print(f"Artinya: {tfidf_matrix.shape[1]} fitur/kata unik (vocabulary)")

    # Menampilkan 10 kata pertama dalam vocabulary yang dipelajari
    print("\nContoh 10 kata pertama dalam vocabulary:")
    print(list(tfidf_vectorizer.get_feature_names_out())[:10])
```

# 2. Word2Vec
Penjelasan Konsep:

Word2Vec adalah pendekatan yang sama sekali berbeda. Ini adalah model neural network yang dilatih untuk "belajar" representasi vektor dari kata-kata berdasarkan konteksnya.

*   Bukan Skor, Tapi Vektor: Tidak seperti TF-IDF yang memberi skor tunggal, Word2Vec mengubah setiap kata menjadi vektor padat (misalnya, [0.12, -0.45, 0.88, ..., 0.34]) dengan 100 dimensi atau lebih.
*   Kekuatan Konteks: Model ini dilatih dengan melihat "jendela" kata-kata (misalnya, 5 kata sebelum dan 5 kata sesudah). Ia belajar bahwa kata-kata yang muncul dalam konteks yang mirip (seperti "sawit" dan "kebun") harus memiliki vektor yang letaknya berdekatan di ruang matematis.
*   Makna Semantik: Hasilnya, vektor ini menangkap makna semantik. Contoh paling terkenal: vektor('Raja') - vektor('Pria') + vektor('Wanita') akan menghasilkan vektor yang sangat dekat dengan vektor('Ratu').




```python
from gensim.models import Word2Vec

import warnings
warnings.filterwarnings('ignore', category=UserWarning, module='gensim')

# 1. Menyiapiapkan input untuk Word2Vec
# Word2Vec butuh input berupa list of list of tokens, misal:
# [ ['kopdes', 'merah', 'putih', 'garap'], ['ganoderma', 'kembang', 'biak'] ]
try:
    tokenized_corpus = [doc.split() for doc in corpus]
    print(f" sudah di-tokenisasi untuk Word2Vec (total {len(tokenized_corpus)} dokumen).")

    # 2. Latih model Word2Vec
    # vector_size=100: Setiap kata akan jadi vektor 100 dimensi
    # window=5: Melihat 5 kata sebelum dan 5 kata sesudah
    # min_count=2: Mengabaikan kata yang hanya muncul 1 kali (mengurangi noise)
    # workers=4: Menggunakan 4 core CPU untuk mempercepat training
    w2v_model = Word2Vec(sentences=tokenized_corpus,
                         vector_size=100,
                         window=5,
                         min_count=2,
                         workers=4)

    w2v_model.save("word2vec.model") # Simpan model
    print("Model Word2Vec berhasil dilatih dan disimpan.")

    # 3. Tampilkan contoh embedding
    print("\n## Contoh Hasil Embedding Word2Vec ##")

    # Coba cek embedding untuk kata "sawit"
    try:
        vector_sawit = w2v_model.wv['sawit']
        print(f"Vektor untuk kata 'sawit' (100 dimensi):")
        print(f"{vector_sawit[:5]}...") # Tampilkan 5 dimensi pertama
    except KeyError:
        print("Kata 'sawit' tidak ditemukan di vocabulary (mungkin karena min_count).")

    # Coba cek kata-kata yang mirip secara semantik
    try:
        similar_words = w2v_model.wv.most_similar('sawit', topn=5)
        print(f"\nKata-kata yang paling mirip dengan 'sawit':")
        print(similar_words)
    except KeyError:
        print("Kata 'sawit' tidak ditemukan di vocabulary.")

except Exception as e:
    print(f"Terjadi error saat melatih Word2Vec: {e}")
```

```python
import pandas as pd

if 'df' not in locals() or 'text_cleaned' not in df.columns:
    print("Error: 'df' atau 'text_cleaned' tidak ditemukan.")
    print("Silakan jalankan ulang kode dari Bagian 2 terlebih dahulu.")
else:
    print("Data 'text_cleaned' ditemukan. Siap untuk pelabelan.")

# ---------------------------------------------------------------------------
# Tentukan Kata Kunci (Lexicon)
# ---------------------------------------------------------------------------
# PENTING: Daftar ini harus disesuaikan dengan konteks "Kelapa Sawit"
# Ini adalah kata-kata dasar (hasil stemming)

# Kata-kata yang cenderung positif dalam konteks ekonomi sawit
positive_words = [
    'dukung', 'positif', 'untung', 'sejahtera', 'ekspor', 'stabil',
    'bantu', 'bangga', 'surplus', 'investasi', 'baik', 'bagus',
    'lowong', 'kerja', 'tumbuh', 'ekonomi'
]

# Kata-kata yang cenderung negatif dalam konteks lingkungan/sosial sawit
negative_words = [
    'rusak', 'hutan', 'bakar', 'asap', 'gundul', 'negatif', 'demo',
    'konflik', 'masalah', 'ilegal', 'curi', 'monopoli', 'gambut',
    'limbah', 'cemar', 'protes', 'korupsi', 'hama', 'ganoderma'
]

# ---------------------------------------------------------------------------
# Membuat Fungsi Pelabelan
# ---------------------------------------------------------------------------
def label_sentiment(text):
    if not isinstance(text, str):
        return 'neutral'

    text_words = set(text.split()) # Pecah teks menjadi kata-kata unik

    # Cek keberadaan kata kunci
    has_positive = any(word in text_words for word in positive_words)
    has_negative = any(word in text_words for word in negative_words)

    if has_positive and not has_negative:
        return 'positive'
    elif not has_positive and has_negative:
        return 'negative'
    else:
        # Jika 'netral', 'positif' & 'negatif' ada, atau tidak ada sama sekali
        return 'neutral'

# ---------------------------------------------------------------------------
# Terapkan Fungsi ke DataFrame
# ---------------------------------------------------------------------------
print("\nMemulai proses pelabelan berbasis aturan...")
# Hapus label palsu jika ada
if 'label' in df.columns:
    df = df.drop(columns=['label'])

df['label'] = df['text_cleaned'].apply(label_sentiment)
print("Pelabelan selesai.")

# ---------------------------------------------------------------------------
# Tampilkan Hasil (VERSI PERBAIKAN)
# ---------------------------------------------------------------------------
print("\n--- Hasil Pelabelan Sentimen ---")
print(df['label'].value_counts())

print("\n--- Contoh Tweet 'positive' ---")
# Menampilkan 2 sampel tweet yang dilabeli 'positive'

positive_samples = df[df['label'] == 'positive'][['full_text', 'label']].head(2)

pd.set_option('display.max_colwidth', None)
display(positive_samples)

print("\n--- Contoh Tweet 'negative' ---")
# Menampilkan 2 sampel tweet yang dilabeli 'negative'
negative_samples = df[df['label'] == 'negative'][['full_text', 'label']].head(2)

display(negative_samples)
```
